### Nearest Neighbors

**Definition**: Nearest neighbors is an algorithm to find class of a data point by majority voting of its nearest neighbors' classes.
The illustration goes like,

![title](../img/knn.png)

```python
import numpy as np
import tensorflow as tf

# Import MINST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
```

```python
# In this example, we limit mnist data
Xtr, Ytr = mnist.train.next_batch(5000) #5000 for training (nn candidates)
Xte, Yte = mnist.test.next_batch(200) #200 for testing

# tf graph Input
xtr = tf.placeholder("float", [None, 784])
xte = tf.placeholder("float", [784])

# Nearest Neighbor calculation using L1 Distance sum ((x_1 - x_2)^2)
distance = tf.reduce_sum(tf.abs(tf.add(xtr, tf.negative(xte))), reduction_indices=1)

# closest point according to distance
pred = tf.argmin(distance, 0)

# initialize the variables
init = tf.global_variables_initializer()

```

```python
accuracy = 0.
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(len(Xte)):
        point_index = sess.run(pred, feed_dict = {xtr: Xtr, xte: Xte[i,:]})
        # finding which index in the range of input rows is closest to the test data point
        print("index: " + str(i) + ", predicted_label: " + str(np.argmax(Ytr[point_index])) + ", actual_label: " + str(np.argmax(Yte[i])))
        if np.argmax(Ytr[point_index]) == np.argmax(Yte[i]):
            accuracy += 1.0/len(Xte)
print('done with accuracy ' + str(accuracy))      
```
