```python
from __future__ import print_function
# Import MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
# ndarray
image = mnist.test.images[1].reshape([28,28])
plt.imshow(image, cmap=plt.get_cmap('gray_r'))
plt.show()
print(mnist.test.labels[1].argmax(axis=0))
```

```python
# train params
display_step = 500
learning_rate = 0.1
epochs = 1000
batch_size = 100

# dimensions 
input_layer = 784
hidden_layer1 = 255
hidden_layer2 = 255
num_classes = 10

# defining placeholders (placeholders are for variables whose value will not be changed by the function)
X_val = tf.placeholder("float", shape=[None, input_layer])
Y_val = tf.placeholder("float", shape=[None, num_classes])
```

```python
# defining the weights and biases
weights = {
    'h1' : tf.Variable(tf.random_normal([input_layer, hidden_layer1])),
    'h2' : tf.Variable(tf.random_normal([hidden_layer1, hidden_layer2])),
    'out' : tf.Variable(tf.random_normal([hidden_layer2, num_classes]))
}
biases = {
    'b1' : tf.Variable(tf.random_normal([hidden_layer1])),
    'b2' : tf.Variable(tf.random_normal([hidden_layer2])),
    'out' : tf.Variable(tf.random_normal([num_classes]))
}
```

```python
# defining the neural net
def neural_net(input_vals):
    layer1 = tf.add(tf.matmul(input_vals, weights['h1']), biases['b1'])
    layer2 = tf.add(tf.matmul(layer1, weights['h2']), biases['b2'])
    out_layer = tf.add(tf.matmul(layer2, weights['out']), biases['out'])
    return out_layer
```

```python
input_fn = neural_net(X_val)
loss_fn = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=input_fn, labels=Y_val))
optimizer_fn = tf.train.AdamOptimizer(learning_rate=learning_rate)
train_fn = optimizer_fn.minimize(loss_fn)
match_preds_fn = tf.equal(tf.argmax(Y_val,1), tf.argmax(input_fn, 1))
accuracy_fn = tf.reduce_mean(tf.cast(match_preds_fn, tf.float32))
```

```python
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(1, epochs+1):
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        sess.run(train_fn, feed_dict={X_val: batch_x, Y_val: batch_y})
        if step%display_step == 0:
            loss_val, acc_val = sess.run([loss_fn, accuracy_fn], feed_dict={X_val: batch_x, Y_val: batch_y})
            print("Epoch " + str(epoch) + ", Current Batch Loss = " + \
                  "{:.4f}".format(loss_val) + ", Current Training Accuracy = " + \
                  "{:.3f}".format(acc_val))
    
    print("Training Finished!")

    # accuracy for MNIST test images
    print("Testing Accuracy:", sess.run(accuracy_fn, feed_dict={X_val: mnist.test.images, Y_val: mnist.test.labels}))
```
